```python
from datetime import datetime

# Function to combine and format a datetime object
def combine_datetime(row):
    date_str = row['Date']
    time_str = row['Time']
    # Combining Date and Time into one string
    datetime_str = f"{date_str} {time_str}"
    # Parsing the string to datetime object
    datetime_obj = datetime.strptime(datetime_str, "%m/%d/%Y %H:%M")
    # Formatting the datetime object
    formatted_datetime = datetime_obj.strftime("%Y/%m/%d %I:%M:%S %p")
    return formatted_datetime

# Applying the function to create the 'Incident Datetime' column
crime_2003_2018_data['Incident Datetime'] = crime_2003_2018_data.apply(combine_datetime, axis=1)
```

```python
import geopandas as gpd
from shapely import wkt
from shapely.geometry import Point

# Importing a CSV containing the latitude and longitude bounding boxes for different neighborhoods

neighborhood_gps_df = pd.read_csv('/content/drive/MyDrive/DATA 240/Datasets/San_Francisco_Analysis_Neighborhoods.csv')

# Convert the 'the_geom' column to actual geometric objects
neighborhood_gps_df['geometry'] = neighborhood_gps_df['the_geom'].apply(wkt.loads)

# Converting the dataFrame to a GeoDataFrame
neighborhoods_gdf = gpd.GeoDataFrame(neighborhood_gps_df, geometry='geometry')

# Creating a spatial index for the neighborhoods
neighborhoods_gdf_sindex = neighborhoods_gdf.sindex
```

```python
# Creating a "Neighborhood Name" column based on the "X" and "Y" columns, and the "neighborhood_gps_df" dataframe

gdf_points = gpd.GeoDataFrame(crime_2003_2018_data, geometry=gpd.points_from_xy(crime_2003_2018_data.X, crime_2003_2018_data.Y))
gdf_points['geometry'] = gdf_points.apply(lambda row: None if pd.isna(row['X']) or pd.isna(row['Y']) else row['geometry'], axis=1)

# Ensuring that the coordinate reference systems match
gdf_points.crs = neighborhoods_gdf.crs

# Using spatial join to find the nearest neighborhood for each point
nearest_neighborhoods = gpd.sjoin_nearest(gdf_points, neighborhoods_gdf[['geometry', 'nhood']], how='left')

# Combining the results back into the original dataFrame
crime_2003_2018_data['Neighborhood Name'] = nearest_neighborhoods['nhood']

# Updating 'Neighborhood Name' to NaN where either 'X' or 'Y' is NaN
crime_2003_2018_data['Neighborhood Name'] = crime_2003_2018_data.apply(lambda row: pd.NA if pd.isna(row['X']) or pd.isna(row['Y']) else row['Neighborhood Name'], axis=1)
```

# **PREPARING THE EXTERNAL DATASETS TO BE MERGED WITH THE INCIDENT DATA**

```python
# Inflation Data

# Load the data from the Excel sheet
file_path = 'Inflation-data.xlsx'
sheet_name = 'hcpi_m'
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Filter the data to include only the row for the United States
us_data = data[data['Country'] == 'United States'].copy()

# Filtering out the non-CPI columns and keeping only the CPI data
cpi_data_only = us_data.iloc[0, 5:-4]

# Converting the index to strings for proper slicing
cpi_data_only_str_index = cpi_data_only.rename(lambda x: str(x))

# Converting the index to datetime format for easier handling
dates = pd.to_datetime(cpi_data_only_str_index.index, format='%Y%m')

# Calculating the year-over-year inflation rate
inflation_rates_yoy = (cpi_data_only_str_index.pct_change(12) * 100).dropna()

# Creating a DataFrame to store the results
inflation_data_yoy = pd.DataFrame({
    'Date': dates[12:],  # Starting from the 13th date since the first 12 months won't have an inflation rate
    'Inflation Rate': inflation_rates_yoy.values
})

# Filtering the results to include data from January 2003 onwards
inflation_data_yoy = inflation_data_yoy[inflation_data_yoy['Date'] >= '2003-01-01']

inflation_data_yoy.to_csv("Inflation Data Monthly - Cleaned.csv", index=False)
```

```python
# Housing price data

df = pd.read_csv("City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")

# Filter the dataframe for San Francisco
sf_data = df[df['RegionName'] == 'San Francisco']

# Extracting the Date and Median Housing Price information
sf_data_melted = sf_data.melt(id_vars=["RegionName"], var_name="Date", value_name="Median Housing Price")
sf_data_melted = sf_data_melted.drop(columns=["RegionName"])
sf_data_melted['Date'] = pd.to_datetime(sf_data_melted['Date'], errors='coerce')
sf_data_melted = sf_data_melted.dropna().reset_index(drop=True)

sf_data_melted.to_csv("Housing Data SF Monthly - Cleaned.csv", index=False)
```

# **FURTHER CLEANING THE FULL MERGED DATASET**

```python
crime_sf_merged = pd.read_csv('/content/drive/MyDrive/DATA 240/Datasets/Crime_SF_Merged.csv')
```

```python
crime_sf_merged['Incident Category'].unique()
```

```python
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Motor Vehicle Theft?', 'Motor Vehicle Theft', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Forgery And Counterfeiting', 'Forgery Counterfeiting', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Drug Offense', 'Drug Narcotic', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Drug Violation', 'Drug Narcotic', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Other Offenses', 'Other', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Other Miscellaneous', 'Other', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Miscellaneous Investigation', 'Other', regex=False)
crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].replace('Suspicious Occ', 'Suspicious', regex=False)

crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].str.replace('/', ' ', regex=False)

# Function to make the first character of each word upper case, and the others lower case

def transform_case(s):
  return ' '.join(word[0].upper() + word[1:].lower() for word in s.split(' '))

crime_sf_merged['Incident Category'] = crime_sf_merged['Incident Category'].apply(transform_case)
```
